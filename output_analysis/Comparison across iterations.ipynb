```python
def plans_cleanup(plans_original):
    
    
    plans=plans_original.copy()
    plans['Mode']=plans['legMode'].shift(-1)
    plans['maxPlanElementIndex']=plans.groupby('personId')['planElementIndex'].transform(max)
    
    
    plans['origin_activityType']=plans['activityType']
    plans['destination_activityType']=plans['activityType'].shift(-2)
    
    
    plans['legTravelTime']=plans['legTravelTime'].shift(-1)
    plans['legDepartureTime']=plans['legDepartureTime'].shift(-1)
    
    plans=plans[plans['planElementIndex']!=plans['maxPlanElementIndex']]
    
    plans_df=plans[plans['activityType'].notnull()]

    return plans_df
```
